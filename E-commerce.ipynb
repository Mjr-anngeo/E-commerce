```python
### In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business. The E-Commerce company has the following rule for computing their own revenue: The company charges
###(i) 25% on the orders having final price (discounted price) greater than 600
###(ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600
###(iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350
###(iv) Otherwise, 5% on the final price (discounted price).
```

```python
### Q 5) The net revenue generated by the E-Commerce company over all orders placed. (10 marks)
```

```python
# Define the function to calculate company revenue
def calculate_revenue(price):
    if price > 600:
        return 0.25 * price
    elif price > 350:
        return 0.15 * price
    elif price > 100:
        return 0.10 * price
    else:
        return 0.05 * price

# Apply the revenue calculation to the dataframe
df['company_revenue'] = df['discounted_price'].apply(calculate_revenue)

# Calculate the total net revenue
total_revenue = df['company_revenue'].sum()

print(f"Total Net Revenue Generated by E-Commerce Company: {total_revenue}")

        
```

### Q6.Calculate the Net BrandRevenue for each Brand and list the top 14 Brand having maximum revenue in descending order (8 marks) 
### Hint: Net BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's net revenue. 

```python
def calculate_deduction(price):# Define the function to calculate E-Commerce company deductions
    if price > 600:
        return 0.25 * price
    elif price > 350:
        return 0.15 * price
    elif price > 100:
        return 0.10 * price
    else:
        return 0.05 * price

# Group by 'Brand' and calculate total discounted_price (BrandRevenue)
brand_revenue = df.groupby('Brand')['discounted_price'].sum().reset_index()
brand_revenue.rename(columns={'discounted_price': 'BrandRevenue'}, inplace=True)

# Apply deductions
brand_revenue['Deduction'] = brand_revenue['BrandRevenue'].apply(calculate_deduction)

# Calculate Net BrandRevenue
brand_revenue['Net BrandRevenue'] = brand_revenue['BrandRevenue'] - brand_revenue['Deduction']

# Sort by Net BrandRevenue in descending order and select top 14
top_brands = brand_revenue.sort_values(by='Net BrandRevenue', ascending=False).head(14)

print(top_brands)
```

### Q 9.Compare performance regionwise
a) Draw a lineplot for the monthly net Revenue of ECommerce Company for each region separately. (4 marks)


```python
sns.lineplot(data=df.groupby(['Region',df['Order_Date'].dt.to_period('M')])['discounted_price'].sum().reset_index().assign(Month_Year=lambda x: x['Order_Date'].astype(str)), x='Month_Year', y='discounted_price', hue='Region', marker='o').set(title='Monthly Net Revenue by Region', xlabel='Month-Year', ylabel='Net Revenue');
plt.xticks(rotation=45)
plt.show()
                
```

### b) Identify the best and the worst performing months for each region. (8 marks) Note: Only those days with actual orders (Order_Date) placed are present in the dataset. Assuming there were no orders on other days

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Extract 'Month-Year'
df['Month-Year'] = df['Order_Date'].dt.to_period('M')

# Group by 'Region' and 'Month-Year', and calculate total revenue
monthly_revenue = df.groupby(['Region', 'Month-Year'])['discounted_price'].sum().reset_index()

# Find the best and worst-performing months for each region
best_worst_months = monthly_revenue.groupby('Region').apply(
    lambda group: pd.DataFrame({
        'Best Month': [group.loc[group['discounted_price'].idxmax(), 'Month-Year']],
        'Best Revenue': [group['discounted_price'].max()],
        'Worst Month': [group.loc[group['discounted_price'].idxmin(), 'Month-Year']],
        'Worst Revenue': [group['discounted_price'].min()]
    })
).reset_index(level=0, drop=True)

print(best_worst_months)
```
